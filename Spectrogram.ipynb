```python
from pathlib import Path
from scipy.io import wavfile
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import signal
from scipy.fft import fftshift
```

```python
def plot_spectrogram(data_name):
    sr, df = wavfile.read(Path.cwd()/"data"/data_name)
    f, t, Sxx = signal.spectrogram(df, sr, return_onesided=True)

    plt.pcolormesh(t, fftshift(f), fftshift(Sxx, axes=0))#, shading='gouraud'

    plt.ylabel('Frequency [Hz]')

    plt.xlabel('Time [sec]')

    plt.show()
```

```python
plot_spectrogram("test_data_2/cavitation/sxmany_RP1_FC_01_C06_SB1_L0_XX_XXXX_YY_YYYY_ZZZZZZ_10.wav")
```

```python
import os
import matplotlib.pyplot as plt

import librosa
import librosa.display

```

```python
def spectro(data_name):
    x, sr = librosa.load(Path.cwd()/"data"/data_name)
                     #"sxmany_200206_007_HKMNetzpumpe04_004_RP1_FC_01_C06_SB1_L0_XX_XXXX_YY_YYYY_ZZZZZZ_10.wav")

    X = librosa.stft(x)
    Xdb = librosa.amplitude_to_db(abs(X))
    plt.figure(figsize=(14, 5))
    librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
    plt.colorbar()
```

```python
spectro("2022_11_25_10_42_44_id_PUMPSMALL003_.wav")
```
